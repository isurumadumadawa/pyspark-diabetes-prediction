```python
import pyspark # only run after findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
data_frame = spark.read.csv('./DatasetForPreProcessing.csv',header=True, inferSchema=True)
data_frame.show()
```

```python
.data_frameprintSchema()
```

```python
clear_data_frame = data_frame.na.drop()
clear_data_frame.show()
```

```python
clear_data_frame = data_frame.na.drop(how="all")
clear_data_frame.show()
```

```python
clear_data_frame = data_frame.na.fill(50)
clear_data_frame.show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=data_frame.columns, 
    outputCols=data_frame.columns,
    ).setStrategy("median")

# Add imputation cols to df
clear_data_frame = imputer.fit(data_frame).transform(data_frame)
clear_data_frame.show()
```
